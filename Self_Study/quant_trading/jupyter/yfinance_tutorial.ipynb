##### 시장 데이터 다운로드

종목의 과거 가격 및 거래량에 대한 데이터를 다운받기 위해서는 `history()`라는 메서드를 사용  
아래는 `history()` 메서드가 가지고 있는 인자(Parameter)와 이에 대한 설명
- `period` : 데이터를 다운받을 기간을 설정합니다. 기본값은 한달(`1mo`). (`1d`, `5d`, `1mo`, `3mo`, `6mo`, `1y`, `2y`, `5y`, `10y`, `ytd`, `max`)
- `interval` : 데이터의 주기를 설정합니다. 주기를 일별보다 낮은 장중으로 설정할 경우 최대 60일간의 데이터만 제공합니다. 기본값은 하루(`1d`). (`1m`, `2m`, `5m`, `15m`, `30m`, `60m`, `90m`, `1h`, `1d`, `5d`, `1wk`, `1mo`, `3mo`)
- `start` : `period` 인자를 사용하지 않을 경우 데이터의 시작일을 설정. `'YYYY-MM-DD'` 문자열을 사용하거나 `datetime` 형식의 날짜를 사용.
- `end` : `period` 인자를 사용하지 않을 경우 데이터의 종료일을 설정. `'YYYY-MM-DD'` 문자열을 사용하거나 `datetime` 형식의 날짜를 사용.
- `prepost` : 시간외거래의 데이터를 결과에 포함시킬것인지를 설정합니다. 기본값은 `False`
- `auto_adjust` : 자동으로 수정주가를 사용할 것인지를 설정합니다. 기본값은 `True`
- `actions` : 주식의 경우 배당이나 주식 분할 같은 기업공시 이벤트를 다운받을지를 설정합니다. 기본값은 `True`

# yfinance로 ETF 가격 데이터 다운로드하기

##### 여러 종목 데이터 한꺼번에 받기

```python
tickers = ['SPY', 'TLT']

etf = yf.Tickers(tickers)

data = etf.history(start='2010-01-01', actions=False)
data
```

```python
data.drop(['Open', 'Low', 'High', 'Volume'], inplace=True, axis=1)
data
```

```python
data = data.droplevel(0, axis=1)
data
```

```python
data.plot()
```

### 입력변수: 수익률, 변동성, 상관계수

#### 1. 수익률

```python
# 수익률 계산
rets = data.pct_change().fillna(0)

# 수익률 데이터 확인
rets.head()
```

```python
# 수익률 데이터 시각화
rets.plot()
plt.title('Historical returns')
plt.xlabel('Date')
plt.ylabel('Returns')
```

#### 2. 변동성

```python
import numpy as np
```

```python
# 전체 기간 변동성 계산
total_vols = rets.std() * np.sqrt(252)
total_vols

# 1년 단위 롤링 변동성 계산 및 연율화
vols = rets.rolling(252).std() * np.sqrt(252)
vols
```

```python
# 변동성 시각화
vols.plot()
plt.title('Historical Vols')
plt.xlabel('Date')
plt.ylabel('Vols')
```

#### 3. 상관계수

```python
# SPY와 TLT의 수익률
spy_rets = rets['SPY']
tlt_rets = rets['TLT']

# 전체 기간 상관계수 계산
total_corr = spy_rets.corr(tlt_rets)
total_corr
```

```python
# 1년 단위 롤링 상관계수 계산
corr = spy_rets.rolling(252).corr(tlt_rets)
corr
```

```python
# 상관계수 시각화
corr.plot()
plt.title('Historical Correlation')
plt.xlabel('Date')
plt.ylabel('Correlation')
```

#### 4. ETF 유니버스 데이터 처리

#### ETF 유니버스

| Ticker | Sector |
| ------ | -----  |
| XLB | 소재 |
| XLE | 에너지 |
| XLF | 금융 |
| XLI | 산업재 |
| XLK | IT |
| XLP | 필수소비재 |
| XLU | 유틸리티 |
| XLV | 헬스케어 |
| XLY | 임의소비재 |

```python
# ETF 유니버스 티커 설정
tickers = ['XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY']

# 티커 객체 선언
yfObj = yf.Tickers(tickers)

# 과거 데이터 다운로드
df = yfObj.history(start='2014-01-01')

# 종가를 제외한 나머지 컬럼 삭제
df.drop(['Open', 'High', 'Low', 'Dividends', 'Stock Splits', 'Volume'], inplace=True, axis=1)

# NaN 값을 이전 값으로 채워넣기
df.ffill(inplace=True)

# 첫 번째 칼럼 레벨 삭제
df = df.droplevel(0, axis=1)

# 주간 단위로 데이터 재설정
df = df.resample('W').last()
```

```python
df
```

```python
# 역사적 가격 그래프 시각화
df.plot()
```

```python
# ETF 수익률 계산
rets = df.pct_change().fillna(0)

# 수익률 시각화
rets.plot(figsize=(12, 8))
```

```python
# 1년 단위 롤링 변동성 계산 및 연율화
vols = rets.rolling(52).std() * np.sqrt(52)

# 변동성 시각화
vols.plot(figsize=(12, 8))
```

```python
# 전체 기간 ETF 상관계수 계산
corr = rets.corr()
corr
```

```python
# seaborn 패키지 임포트
import seaborn as sns

# 히트맵 시각화
plt.figure(figsize=(12,6))
sns.heatmap(corr, annot=True)
```

##### 공분산 행렬 특징 이해하기

```python
# 공분산 계산 및 연율화
cov = rets.cov() * 52
cov
```

```python
# var() 메서드를 사용한 ETF 분산 계산
vols_1 = rets.var() * 52
vols_1
```

```python
# 공분산 행렬 및 diag() 메서드를 사용한 ETF 분산 계산
vols_2 = np.diag(cov)
vols_2
```
